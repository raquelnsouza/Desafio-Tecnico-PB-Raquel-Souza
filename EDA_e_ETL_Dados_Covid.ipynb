```python
# fato Dados Covid

for index, row in fdados_covid.iterrows():
    city_ibge_code = row['city_ibge_code']
    date = row['date']
    epidemiological_week = row['epidemiological_week']
    last_available_confirmed = row['last_available_confirmed']
    last_available_confirmed_per_100k_inhabitants = row['last_available_confirmed_per_100k_inhabitants']
    last_available_date = row['last_available_date']
    last_available_death_rate = row['last_available_death_rate']
    last_available_deaths = row['last_available_deaths']
    order_for_place = row['order_for_place']
    new_confirmed = row['new_confirmed']
    new_deaths = row['new_deaths']
    id_place_type = row['id_place_type']
    id_state = row['id_state']
    id_is_last = row['id_is_last']
    id_is_repeated = row['id_is_repeated']

    query = '''
    INSERT INTO covid.f_dados_covid (
        city_ibge_code, date, epidemiological_week, last_available_confirmed,
        last_available_confirmed_per_100k_inhabitants, last_available_date,
        last_available_death_rate, last_available_deaths, order_for_place,
        new_confirmed, new_deaths, id_place_type, id_state, id_is_last, id_is_repeated
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    '''
    cursor.execute(query, (
        city_ibge_code, date, epidemiological_week, last_available_confirmed,
        last_available_confirmed_per_100k_inhabitants, last_available_date,
        last_available_death_rate, last_available_deaths, order_for_place,
        new_confirmed, new_deaths, id_place_type, id_state, id_is_last, id_is_repeated
    ))
    conn.commit()
```
